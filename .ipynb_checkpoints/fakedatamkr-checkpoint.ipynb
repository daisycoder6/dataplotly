```python
import pandas as pd
import numpy as np
import itertools
```

```python
noisedata = np.random.rand(150,5)
```

```python
noisedata
```

```python
dutid =[['dut1']*15,['dut2']*15,['dut3']*15,['dut4']*15,['dut5']*15,['dut6']*15,['dut7']*15,['dut8']*15,['dut9']*15,['dut10']*15]
```

```python
duts=itertools.chain.from_iterable(dutid)
```

```python
duts
```

```python
ids = list(duts)
```

```python
ids
```

```python
noisedf = pd.DataFrame(noisedata,index=None)
```

```python
noisedf.tail()
```

```python
noisedf['ids']=ids
```

```python
noisedf.head()
```

```python
conditions =[['ecg']*5,['resp']*5,['pace']*5]
conds = list(itertools.chain.from_iterable(conditions))
condsall = conds*10
condsall
```

```python
noisedf['Condition']=condsall
```

```python
noisedf.head()
```

```python
#noisedf.columns=['ids','Conditions','0','1','2','3','4']
```

```python
noisedf=noisedf[['ids','Condition',0,1,2,3,4]]
```

```python
noisedf.columns=['ids','Conditions','ch1','ch2','ch3','ch4','ch5']
```
